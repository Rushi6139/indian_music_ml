```python

track_data = []
for track in results['tracks'][:10]:
    track_data.append([
        track['name'],                            # Name
        track['album']['name'],                   # Album
        track['artists'][0]['name'],              # Artist
        track['album']['release_date'],           # Release Date
        track['duration_ms'] // 1000,             # Length (seconds)
        track['popularity']                       # Popularity score
    ])

# Create DataFrame
df_final = pd.DataFrame(track_data, columns=[
    'Name', 'Album', 'Artist', 'Release_Date', 'Length_sec', 'Popularity'
])

print(df_final)
```

```python
def getTrackIDs(user, playlist_id):
    track_ids = []
    album = sp.user_playlist(user, playlist_id)
    for item in album['tracks']['items']:
        track = item['track']
        track_ids.append(track['id'])
    return track_ids
track_ids = getTrackIDs('spotify', "0MlHYL3NnKMbneksYevshZ")
print(len(track_ids))
print(track_ids)
```

```python
def getPlaylistFeatures(id):
 track_info = sp.track(id)
#Track Info
 name = track_info['name']
 album = track_info['album']['name']
 artist = track_info['album']['artists'][0]['name']
 release_date = track_info['album']['release_date']
 length = track_info['duration_ms']
 popularity = track_info['popularity']
 track_data =[name, album, artist, release_date, length]
 return track_data
```

```python
# Filter songs with Popularity > 70
pop_96 = df_final.loc[df_final['Popularity'] > 70].sort_values(by='Popularity', ascending=False)

# Pie chart of Popularity by Artist
import plotly.express as px
fig = px.pie(
	pop_96,
	values='Popularity',
	names='Artist',
	title='Popularity of Songs Greater than 70 by Artist in this Playlist'
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
